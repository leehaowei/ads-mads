### Option 1

```python
class Option1:
    
    def __init__(self, synonyms, sentence):
        self.syn = synonyms
        self.sen = sentence
        
    def solve(self):
        # part 1
        my_syn = self.syn.copy()
        for i in range(1, len(my_syn)):
            check = my_syn[i]

            for j in range(i):
                before = my_syn[j]
                for c in check:
                    if c in before:
                        before += check
                        my_syn.pop(i)
                        break
                        
        final_syn = [set(_) for _ in my_syn]
        final_syn = [list(_) for _ in final_syn]
        self.syn = final_syn
    
        # part 2
        answers = []
        sentence_list = self.sen.split()
        for each in self.syn[0]:
            if each in sentence_list:
                target1 = each

        for each in self.syn[1]:
            if each in sentence_list:
                target2 = each

        for i in range(len(self.syn[0])):
            for j in range(len(self.syn[1])):
                a = sentence
                a = a.replace(target1, self.syn[0][i])
                a = a.replace(target2, self.syn[1][j])
                answers.append(a)
                
        return answers
```

```python
synonyms = [["happy","joyful"],["sad","sorrow"],["joyful","cheerful"]]
sentence = "I am happy today but was sad yesterday"
```

```python
o1 = Option1(synonyms, sentence)
```

```python
o1.solve()
```

### Option 2

```python
def option2(matrix):
    ans = -1
    first = matrix[0]
    for i in range(len(first)):
        point = 0
        check = first[i]
        for each in matrix[1:]:
            if check in each:
                point += 1
        if point == len(matrix)-1:
            ans = first[i]
            return ans
```

```python
test = [[1,2,3,4,6],[2,4,5,8,10],[3,5,7,9,11],[1,3,5,7,9]]
option2(test)
```
