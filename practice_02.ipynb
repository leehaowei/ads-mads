### 1. Integer to Roman
####  The biggest number we can form in Roman numerals is MMMCMXCIX, or 3,999.

```python
class Solution:
    def to_roman(self, num, ans=''):
        if num == 0:
            return ans
        
        elif num >= 900:
            denom, symbol = (900, 'MD') if num < 1000 else (1000, 'M')
            new_num, new_ans = self.new(num, denom, ans, symbol)
            return self.to_roman(new_num, new_ans)                 
       
        elif num >= 400:
            denom, symbol = (400, 'CD') if num < 500 else (500, 'D')
            new_num, new_ans = self.new(num, denom, ans, symbol)
            return self.to_roman(new_num, new_ans)     
    
        elif num >= 90:
            denom, symbol = (90, 'XC') if num < 100 else (100, 'C')
            new_num, new_ans = self.new(num, denom, ans, symbol)
            return self.to_roman(new_num, new_ans)     
        
        elif num >= 40:
            denom, symbol = (40, 'XL') if num < 50 else (50, 'L')
            new_num, new_ans = self.new(num, denom, ans, symbol)
            return self.to_roman(new_num, new_ans)
       
        elif num >= 9:
            denom, symbol = (9, 'IX') if num < 10 else (10, 'X')
            new_num, new_ans = self.new(num, denom, ans, symbol)
            return self.to_roman(new_num, new_ans)
        
        elif num >= 4:
            denom, symbol = (4, 'IV') if num < 5 else (5, 'V')
            new_num, new_ans = self.new(num, denom, ans, symbol)
            return self.to_roman(new_num, new_ans)
        
        elif num >= 1:
            new_num, new_ans = self.new(num, 1, ans, 'I')
            return self.to_roman(new_num, new_ans)
        
    def new(self, num, denom, ans, symbol):
        new_ans = ans + (num // denom) * symbol
        new_num = num - (num // denom) * denom
        return new_num, new_ans
```

```python
s = Solution()
print(s.to_roman(3794))
```

### 2. Roman to Integer

```python
class Solution:
    def roman_to_int(self, roman, ans=0):
        if not roman:    # when roman becomes empty string
            return ans
        
        check = roman[0]
        
        if check == 'M':
            return self.roman_to_int(roman[1:], ans + 1000)
            
        if check == 'D':
            return self.roman_to_int(roman[1:], ans + 500)
            
        if check == 'C':
            if len(roman) > 1:
                next_ = roman[1]
                new_ans = ans + 900 if next_ == 'M' else ans + 400  # else next_ == 'D'
                return self.roman_to_int(roman[2:], new_ans)
            
            else:
                return self.roman_to_int(roman[1:], ans + 100)
            
        if check == 'L':
            return self.roman_to_int(roman[1:], ans + 50)
            
        if check == 'X':
            if len(roman) > 1:
                next_ = roman[1]
                new_ans = ans + 90 if next_ == 'C' else ans + 40  # else next_ == 'L'
                return self.roman_to_int(roman[2:], new_ans)
            
            else:
                return self.roman_to_int(roman[1:], ans + 10)
            
        if check == 'V':
            return self.roman_to_int(roman[1:], ans + 5)
            
        if check == 'I':
            if len(roman) > 1:
                next_ = roman[1]
                new_ans = ans + 9 if next_ == 'X' else ans + 4  # else next_ == 'V'
                return self.roman_to_int(roman[2:], new_ans)
            
            else:
                return self.roman_to_int(roman[1:], ans + 1)
            
        return ans
```

```python
roman_1 = 'MC'
roman_2 = 'MD'
roman_3 = 'MCM'
roman_4 = 'MCD'
roman_5 = 'MCMXL'
roman_6 = 'MCMXLIX'
roman_7 = 'MCDXCIX'
R = [roman_1, roman_2, roman_3, roman_4, roman_5, roman_6, roman_7]
s = Solution()
```

```python
for r in R:
    print(s.roman_to_int(r))
```

### 3. Maximum Profit from Trading Stocks

1. https://www.geeksforgeeks.org/maximum-profit-by-buying-and-selling-a-share-at-most-k-times/
2. https://leetcode.com/problems/best-time-to-buy-and-sell-stock-iv/solution/

#### Easy version

```python
# Accepted
class Solution_1:
    def maxProfit(self, prices):
        max_profit = 0
        accum_margin = 0
        ln = len(prices)
        for i in range(0, ln-1):
            margin = prices[i+1] - prices[i]
            
            if accum_margin + margin > 0:
                accum_margin += margin
            
            if accum_margin > max_profit:
                max_profit = accum_margin
        
        return max_profit
```

```python
# Accepted
class Solution_2:
    def maxProfit(self, prices):
        max_profit, accum_margin = 0, 0
        ln = len(prices)
        
        for i in range(0, ln-1):
            margin = prices[i+1] - prices[i]    
            accum_margin = max(accum_margin + margin, 0)
            max_profit = max(accum_margin, max_profit)
            
        return max_profit
```

```python
l = [7,1,5,3,6,4]
s1 = Solution_1()
s2 = Solution_2()
```

```python
print(s1.maxProfit(l))
print(s2.maxProfit(l))
```

### 4. Check Word Abbreviation

#### 408. Valid Word Abbreviation

```python
# Time Limit Exceeded
class Solution:
    
    def sol(self, given_s, abbr, ans=[]):
        final = self.abbrev(given_s, current='')
        
        for a in final:
            ans.append(self.revise(a))
            
        if abbr in ans:
            return True
        return False
    
    def revise(self, s, e = '', total = 0, switch = False):
        
        for i in range(len(s)): 
            _ = s[i]
            if _.isdigit():
                switch = True
                total += 1

            else:
                if switch:
                    e += str(total)
                    switch = False
                    total = 0
                e += _

            # OBOB
            if switch and i == len(s)-1:
                e += str(total)
        return e
        
    def abbrev(self, choices, current='', ans = []):
        # Base case
        if len(choices) == 0:
            ans.append(current)
            return ans

        else:
            # Choose and explore
            ele = choices[0]
            choices = choices[1:]
            current += ele
            ans = self.abbrev(choices, current)

            # Not choose and explore
            current = current[:-1] + '1'
            return self.abbrev(choices, current, ans)

            # Un-choose
            choices += ele
```

```python
s = Solution()
s.sol(given_s="internationalization", abbr = "i12iz4n")
```

### 5. Password Checker

```python
def is_strong(password):
    ln, up, low, dig, repeat = 1, 1, 1, 1, 1
    count = 0
    current = None
    
    if 6 <= len(password) <= 20:
        ln -= 1
        
    for i in range(len(password)):
        s = password[i]
        
        if s == current:
            count += 1
        else:
            count = 0
        current = password[i]
    
        if up and not s.isdigit() and s.upper() == s:
            up -= 1
        
        if low and not s.isdigit() and s.lower() == s:
            low -= 1
        
        if dig and s.isdigit():
            dig -= 1
    
    if count < 2:
        repeat -= 1
    return ln + up + low + dig + repeat
```

```python
test1 = 'BBa12CCCDDD'
test2 = 'BBa12CCDD'
test3 = 'bba12ccdd'
test4 = 'BBA12CCDD'
test5 = '1222'
test6 = 'AAA'
tests = [test1, test2, test3, test4, test5, test6]
for t in tests:
    print(f'Require at least {is_strong(t)} changes')
```

### 6. Sliding Window Median

```python
class Solution:
    def find_medians(self, nums, k, ans=[]):
        if len(nums) < k:
            return ans
        
        a_list = sorted(nums[:k])
        ln = len(a_list)
        mid_id = ln // 2
        
        if ln % 2 == 0:
            median = (a_list[mid_id] + a_list[mid_id+1]) / 2
        else:
            median = a_list[mid_id]
        ans.append(median)
        
        return self.find_medians(nums[1:], k, ans)
```

```python
nums =  [1,3,-1,-3,5,3,6,7]
k = 3
s = Solution()
print(s.find_medians(nums, k))
```
